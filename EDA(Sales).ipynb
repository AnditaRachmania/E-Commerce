### Describe dan handling missing value (jika ada)

```python
desc = []

for i in df.columns:
    desc.append([
        i,
        df[i].dtypes,
        df[i].isnull().sum(),
        round((((df[i].isnull().sum())/(len(df)))*100), 2),
        df[i].nunique(),
        df[i].drop_duplicates().sample(2).values
    ])
```

```python
dfDesc = pd.DataFrame(desc, columns=['dataFeatures', 'dataType', 'null', 'nullPercentage', 'unique', 'uniqueSample'])
```

```python
dfDesc
```

Dari output Tabel dfDesc, terlihat bahwa dalam dataset, tiap kolom tidak terdapat value null

## Analisa Kolom City

```python
plt.figure(figsize=(20, 10))
sns.countplot(y='City', data=df, palette='rainbow', order=df['City'].value_counts().index)
plt.grid(True)
plt.title("Record Penjualan per Kota")
plt.show()
```

Grafik di atas menunjukkan jumlah penjualan e-commerce tiap kota di India. Dari grafik terlihat bahwa penjualan mencapai angka tertinggi di Kota Indore, diikuti oleh Mumbai dan Delhi.

Indore merupakan salah satu kota metropolitan terbesar di India bagian tengah, yang merupakan kawasan terpadat dan kota terbesar dari State Mdhya Pradesh (sumber: https://en.wikipedia.org/wiki/Indore). Sementara Mumbai adalah Ibu Kota state Maharashtra, yang pada tahun 2018 merupakan kota terpadat kedua di India setelah Delhi (sumber: https://en.wikipedia.org/wiki/Mumbai)

```python
kota = df[(df['City'] == "Mumbai") | (df['City'] == "Indore") | (df['City'] == "Pune")]
kota
```

```python
sns.countplot(y='Category', data=kota, palette='coolwarm')
plt.title("Penjualan E-Commerce di Kota Mumbai, Indore, dan Pune", fontsize=15)
```

Dari ketiga kota dimana terdapat record penjualan yang terbanyak, kategori barang yang paling banyak terjual adalah Clothing, sementara penjualan kategori electronics dan furniture hampir sama besarnya.
